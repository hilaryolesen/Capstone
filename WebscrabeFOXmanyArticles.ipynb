```python
#this and following 2 cells: access webpages, acquire html, put in list
pages=[]
for URL in URLs:
    page = requests.get(URL)
    if page.status_code!=200:
        print(URL,page.status_code)
    else:
        pages.append(page)
```

```python
HTMLstrs=[]
for page in pages:
    HTMLstrs.append(page.text)
```

```python
soups = []
for HTMLstr in HTMLstrs:
    soups.append(BeautifulSoup(HTMLstr, "lxml"))
```

```python
#gather titles from each webpage
titles=[]
for soup in soups:
    titles.append(soup.title.string)
```

```python
#gather authors from each webpage
authors=[]
for soup in soups:
    authors.append(soup.find("meta", {"name" : "parsely-author"})["content"])
```

```python
#pull article text, customized for website to avoid ads and extras, and to remove whitespace
txts=[]
for soup in soups:
    txt=''
    for i in soup.find_all("p"):
        txt=txt+" "+i.getText()
    if " SEE MORE:" in txt:
        cutoff=txt.find(" SEE MORE:")
    else:
        cutoff=len(txt)
    txts.append(txt[6:cutoff])
```

```python
#test print first text in list
print(txts[0])
```

```python
#gather and format dates by last modified, avoiding one strangely-formatted date
artdates=[]
for soup in soups:
    try:
        artdate=datetime.fromisoformat(soup.find("meta", {"property" : "article:modified_time"})["content"].split(".")[0])
        for i in soup.find_all("meta", {"property" : "article:modified_time"}):
            artdate=datetime.fromisoformat(i["content"])
    except:
        artdate=datetime.fromisoformat(soup.find("meta", {"property" : "article:published_time"})["content"].split(".")[0])
    artdates.append(artdate)
```

```python
#turn lists into dataframe, assign same source to all
df=pd.DataFrame({'Titles':titles,'Dates':artdates,'Authors':authors,'URLs':URLs,'Texts':txts})
df['Source']="FOX 4"
```
