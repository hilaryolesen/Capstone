```python
#this and following 2 cells: access webpages, acquire html, put in list
pages=[]
for URL in URLs:
    page = requests.get(URL)
    if page.status_code!=200:
        print(URL,page.status_code)
    else:
        pages.append(page)
```

```python
HTMLstrs=[]
for page in pages:
    HTMLstrs.append(page.text)
```

```python
soups = []
for HTMLstr in HTMLstrs:
    soups.append(BeautifulSoup(HTMLstr, "lxml"))
```

```python
#gather titles from each webpage
titles=[]
for soup in soups:
    titles.append(soup.title.string)
```

```python
#gather authors from each webpage
authors=[]
for soup in soups:
    authors.append(soup.find("meta", {"name" : "author"})["content"])
```

```python
#pull article text, customized for website to avoid ads and extras, and to remove whitespace
txts=[]
for soup in soups:
    txt=''
    for i in soup.find_all("p"):
        txt+=i.getText()
    cutoff=txt.find("\n\n\n\n\t\t\t\n\nSubmit a correction")
    txts.append(txt[16:cutoff].replace("\n\n\n"," ").replace("\n\n"," ").replace("\n"," "))
```

```python
#test print first text in list
print(txts[0])
```

```python
#gather and format dates by last modified, avoiding timezone information
artdates=[]
for soup in soups:
    try:
        artdate=datetime.fromisoformat(soup.find("meta", {"property" : "article:modified_time"})["content"][:-6])
        for i in soup.find_all("meta", {"property" : "article:modified_time"}):
            artdate=datetime.fromisoformat(i["content"][:-6])
    except:
        artdate=datetime.fromisoformat(soup.find("meta", {"property" : "article:published_time"})["content"][:-6])
    artdates.append(artdate)
```

```python
#turn lists into dataframe, assign same source to all
df=pd.DataFrame({'Titles':titles,'Dates':artdates,'Authors':authors,'URLs':URLs,'Texts':txts})
df['Source']="WINK News"
```
